```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Start with just one resonant frequency
#Set parameters

omega_p = 1
omega_0 =10

#axis range - adjust if needed
omega = np.arange(0.,30,0.1)

#calculate real and imaginary parts, then put together as n and kappa
real_chi = (omega_0**2-omega**2)/((omega_0**2-omega**2)**2 + omega**2)
imag_chi = omega/((omega_0**2-omega**2)**2 + omega**2)
    
n = np.sqrt((1+real_chi + np.sqrt((1+real_chi)**2+imag_chi**2))/2)
kappa = imag_chi/2/n
    
#plot them
plt.plot(omega,n,'b',omega,kappa,'r')
plt.show()
```

```python
#Start with just one resonant frequency
#Set parameters

omega_p = 1
omega_0 =10


#axis range - adjust if needed
omega = np.arange(0,30,.1)

#calculate real and imaginary parts, then put together as n and kappa

real_chi = omega_p**2*(omega_0**2-omega**2)/((omega_0**2-omega**2)**2+omega**2)
imag_chi = omega_p**2*omega/((omega_0**2-omega**2)**2+omega**2)
    
n = np.sqrt((1+real_chi+np.sqrt((1+real_chi)**2+imag_chi**2))/2)
kappa = imag_chi/2/n
    
#plot them
plt.plot(omega,n,'b',omega,kappa,'r')
plt.show()
```

```python
#Set parameters
#Now there are matching-length arrays of parameters - one for each omega_0
fs = [1,0.25,0.25, .25]
omega_ps = [5,5,3,5]
omega_0s = [10.,15.,25., 27.]

#axis range - adjust if needed, based on omega_0s
omega = np.arange(0,30,.1)
real_chi = np.zeros_like(omega)
imag_chi = np.zeros_like(omega)

#calculate real and imaginary parts for each omega_0 and sum over them

for i,omega_0 in enumerate(omega_0s):
    f=fs[i]
    omega_p = omega_ps[i]
    real_chi = real_chi + f*omega_p**2*(omega_0**2-omega**2)/((omega_0**2-omega**2)**2+omega**2)
    imag_chi = imag_chi + f*omega_p**2*omega/((omega_0**2-omega**2)**2+omega**2)

#calculate n and kappa 
n = np.sqrt((1+real_chi+np.sqrt((1+real_chi)**2+imag_chi**2))/2)
kappa = imag_chi/2/n
    
#plot them
plt.plot(omega,n,'b',omega,kappa,'r')
plt.show()
```
